## Data munging
***

1. Dropping some features
2. Changing the date recorded into three different features for year, month and day
3. Changing string valued (categorical) features into one hot vectors 

```python
columns_to_drop = ['id','funder', 'installer','wpt_name',
           'num_private','basin','subvillage','region','region_code','district_code','lga','ward','public_meeting', 'recorded_by',
        'scheme_management','scheme_name', 'extraction_type_group','extraction_type_class']
data_PumpItUp.drop(columns_to_drop, axis=1, inplace=True)
```

```python
data_PumpItUp['date_recorded'] = pd.to_datetime(data_PumpItUp['date_recorded'])
data_PumpItUp['year_recorded'] = data_PumpItUp['date_recorded'].map(lambda x: x.date().year)
data_PumpItUp['month_recorded'] = data_PumpItUp['date_recorded'].map(lambda x: x.date().month)
data_PumpItUp['day_recorded'] = data_PumpItUp['date_recorded'].map(lambda x: x.date().day)
data_PumpItUp = data_PumpItUp.drop('date_recorded', 1)
```

```python
import numpy
from sklearn.feature_extraction import DictVectorizer


def one_hot_dataframe(data, cols, replace=False):
    """ Takes a dataframe and a list of columns that need to be encoded.
        Returns a 3-tuple comprising the data, the vectorized data,
        and the fitted vectorizor.
    """
    vec = DictVectorizer()
    mkdict = lambda row: dict((col, row[col]) for col in cols)
    vecData = pd.DataFrame(vec.fit_transform(data[cols].apply(mkdict, axis=1)).toarray())
    vecData.columns = vec.get_feature_names()
    vecData.index = data.index
    if replace is True:
        data = data.drop(cols, axis=1)
        data = data.join(vecData)
    return (data, vecData, vec)
```

```python
colums_to_one_hot_vector = ['permit', 'extraction_type','management','management_group',
                                                        'source','source_type','source_class','waterpoint_type','waterpoint_type_group',
                                                        'payment', 'payment_type','water_quality','quantity','quality_group', 'quantity_group']
```

```python
data_PumpItUp, _, _ = one_hot_dataframe(data_PumpItUp, colums_to_one_hot_vector, replace=True)
```

### Some final modifications
--------------------
The feature **'extraction_type=other - mkulima/shinyanga'** is removed since it is not present in the submission data. If it is not removed, the number of features for the trained model and the submission data will not be the same. The problem can be solve otherwise by first merging all the data before the treatment.


```python
data_PumpItUp = data_PumpItUp.loc[data_PumpItUp['extraction_type=other - mkulima/shinyanga'] ==0.0].drop(['extraction_type=other - mkulima/shinyanga'], axis=1)
```

A class imbalance can be seen. the "functional needs repair" class is underrepresented. 

```python
ax = sns.countplot(x="status_group", data=data_PumpItUp)
```

```python
data_PumpItUp = data_PumpItUp.fillna(data_PumpItUp.mean())
```
